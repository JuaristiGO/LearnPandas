# EDA Exploratory Data Analysis


Look for patterns, mistakes, missing values, etc.

```python
# Import seaborn and matplotlyb to visualize data as we go, to comprehend how data is going.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\Gerardo\Downloads\world_population.csv')
df
```

```python
# .info() gives you highlevel information (nulls, data types, columns)
df.info()
```

```python
# .describe() allows you to get an overview of all columns values.
df.describe()
```

```python
# to change scientific notation use .set_option('display')
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
# Null values check and count them with .isnull().sum()
df.isnull().sum()
```

```python
# Use of .nunique() to look for unique values and compare it with the logic of the dataset
df.nunique()
```

```python
# Begin with exploration using sort_values()
df.sort_values(by = '2022 Population', ascending = False).head(10)
```

```python
# Look for correlation in numeric values to compare them and look how close correlated they are use .corr()
corr = df.corr()
sns.heatmap(corr, annot = True)
plt.rcParams['figure.figsize'] = (10,7)
plt.show()
```

```python
# Group columns with .groupby() to have a closer look to our data
df.groupby('Continent').mean().sort_values('2022 Population', ascending = False)
```

```python
# Check for Oceania continent
df[df['Continent'].str.contains('Oceania')]
```

```python
# visualize growth population per time using transpose and filtering only needed columns
df2 = df.groupby('Continent')[df.columns[5:13]].mean().sort_values('2022 Population', ascending = False)
df_plot = df2.transpose()
```

```python
plt.rcParams['figure.figsize'] = (20,7)
df_plot.plot()
```

```python
# Create a boxplot to have an overview and check the out the outlier dots
df.boxplot(figsize=(20,10))
```

```python
# Check dtypes of the df 
df.dtypes
```

```python
# filter ftypes to check the data types and the objects they have.
df.select_dtypes(include='object')
df.select_dtypes(include='float')
df.select_dtypes(include='int')
```
