```python
# Filter a single column
df['Name']
```

```python
# Filter many columns
df[['Name', 'Industry', 'Employees']]
```

```python
# Filter columns with operators and change dtypes 
df['Employees'] = df['Employees'].str.replace(',','')
df['Revenue (USD millions)'] = df['Revenue (USD millions)'].str.replace(',','')
df = df.astype({'Revenue (USD millions)':'int','Employees':'int'})
df[df['Employees']>50000]
```

```python
# Filter more than 1 column with operators
df[(df['Employees']>50000) & (df['Revenue (USD millions)'] >300000)]
```

```python
# Use of .isin() works like an IN SQL clause select a column an the records you want
specific_companies = ['Walmart','Amazon','Apple','Microsoft','Chevron Corporation','Costco','PBF Energy','Tesla']
df[df['Name'].isin(specific_companies)]
```

```python
# Use of contain() works like a LIKE clause from SQL works with str
df[df['Name'].str.contains('on')]
```

```python
# Filter by index with filter loc iloc
# Set an index with .set_index()
df2 = df.set_index('Name')
# Filter from the index with filter() function
df2.filter(items = ['Industry','Employees'])

```

```python
# Can use filter with like arg and axis 0 to look up for the records
df2.filter(like = 'Corporation', axis = 0)
```

```python
# Use of .loc function to filter bring the whole location of the item inside loc
df2.loc['Apple']
```

```python
# Use of .iloc function works like loc but with index number
df2.iloc[7]
```

```python
# .sort_values() function order and filter fuctions as wanted
df[df['Employees']>40000].sort_values(by='Employees')
```

```python
# Multiple Indexes with a list to add more than 1 column.
df = df.set_index(['Industry', 'Name','Rank'])
df
```

```python
# Sort Index alphabetically with sort_index()
df = df.sort_index()
df
```

```python
# Search by multiple indexes with loc[]
df.loc['Petroleum industry','Chevron Corporation']
```

```python
# Group By in pandas with .groupby() on Industry column
df.reset_index(inplace = True)
df.set_index('Rank',inplace = True)
df = df.sort_index()
df_group_by_frame = df.groupby('Industry')
df_group_by_frame 
```

```python
# Add aggregations to the group by function in order to get an output like mean() as an aggregate function.
df_group_by_frame.mean('Revenue (USD millions)')
```

```python
# Most common aggregate functions .count()
#df.groupby('Industry').count()
#df.groupby('Industry').min('Revenue (USD millions)')
#df.groupby('Industry').max('Revenue (USD millions)')
#df.groupby('Industry').sum('Revenue (USD millions)')
```

```python
# Aggregation function .agg() to have multiple aggs in the same groupby statement

df.groupby('Industry').agg({'Revenue (USD millions)':['mean','max','min','sum'], 'Employees':['mean','max','min','sum'] })
```

```python
df.groupby(df['Industry']).agg({'Revenue (USD millions)':['mean','max','min','sum'], 'Employees':['mean','max','min','sum'] })
```

```python
# Groupby multiple columns
df.groupby(['Industry', 'Headquarters']).describe()
```
