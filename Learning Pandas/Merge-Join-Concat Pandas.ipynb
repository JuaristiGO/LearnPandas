# MERGE JOIN AND CONCATENATE

```python
import pandas as pd
```

```python
df1 = pd.read_csv(r'C:\Users\Gerardo\Downloads\NHL_1992_standings.csv')
df2 = pd.read_csv(r'C:\Users\Gerardo\Documents\Python\Python Web Scraping\NHL 90s\NHL_1993_standings.csv')
```

```python
# Merge the most important kind of join (like SQL joins) df1.merge(df2)
df1.merge(df2, on='Team Name')
```

```python
# Left/Right join with .merge() argument how = left, custom x and y values to identify each
df1.merge(df2, how = 'right', on='Team Name', suffixes = (' 1992',' 1993'))
```

```python
# Outer Join with merge() how = 'outer' 
df1.merge(df2, how='outer',on='Team Name')
```

```python
# Join function in pandas they work with indexes so have to be specified
df3 = df1.set_index('Team Name').join(df2.set_index('Team Name'), lsuffix = ' 1992', rsuffix = ' 1993', how = 'outer')
df3
```

```python
# Use of concatenate df with .concat() works like putting one data frame on top of the other rather than next to one-another
pd.concat([df1,df2])
```

```python
# Append function .append() to add new values to the df
df1.append(df2)
```
