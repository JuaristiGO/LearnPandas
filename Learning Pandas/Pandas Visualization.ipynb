# Pandas Visualization

```python
import pandas as pd
from matplotlib.pyplot import matplotlib as plt
import numpy as np
```

```python
df = pd.read_csv(r'C:\Users\Gerardo\Documents\Python\Python Web Scraping\NHL 90s\NHL_1993_standings.csv')
df
```

```python
# Clean dataset and set index
df['Rank'] = [i+1 for i,row in enumerate(df['Team Name'])]
df.set_index('Rank', inplace = True)
```

```python
df.drop('OT Losses', axis = 1, inplace = True)
df
```

```python
# Use general .plot() function x and y labels 1st and 2nd arg, kind = for the visualization plot etc.
df.plot('Team Name', 'Wins', kind = 'bar', ylabel = 'Wins')
```

```python
# Horizontal bar plot
df.plot.barh('Team Name','Wins')
```

```python
# Scatter plot specify x and y axis
df.plot.scatter('Wins','Goals For (GF)', s = 40)
```

```python
df.plot.scatter('Wins','Goals Against (GA)', s = 40, c = 'red')
```
